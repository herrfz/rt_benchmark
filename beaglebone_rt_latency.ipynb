### cyclictest

Command to run test:

No real-time:

```
# /usr/local/sbin/cyclictest -l100000000 -m -Sp99 -i200 -h500 -n -q
```

PREEMPT_RT:

```
# /usr/local/sbin/cyclictest -l100000000 -m -Sp99 -i200 -h500 -n -q
```

Xenomai:

```
# /usr/xenomai/bin/cyclictest -l100000000 -p99 -i200 -h500 -n -q
```

Note that the `cyclictest` program is different for Xenomai, it is adapted to use its real-time APIs.

Each test is run with a background load pushing the CPU close to 100%:

```
# /usr/local/sbin/hackbench -l100000000
```

What it does is that it creates 400 tasks, each sending 100000000 messages of 100 bytes.

The program outputs a 'histogram' data, i.e. each line consists of a latency value and a count of that observed latency value during the test. Additionally, some summary data is output as well:

* minimum latency
* average lateny
* maximum latency
* number of histogram overflows: how many times a measured latency exceeds the specified maximum histogram value
* etc.

````python
# load as pandas DataFrame
df1 = pd.read_csv('files/nort_clean.hist', sep=' ', header=None, names=['latency', 'no_rt'])
df2 = pd.read_csv('files/rt_clean.hist', sep=' ', header=None, names=['latency', 'preempt_rt'])
df3 = pd.read_csv('files/rt_xenomai_clean.hist', sep=' ', header=None, names=['latency', 'rt_xenomai'])
````

````python
# DataFrame wrangling to get the shape required for plotting
merged = df1.merge(df2, how='outer', on='latency').merge(df3, how='outer', on='latency')
melted = pd.melt(merged, id_vars='latency')
melted.tail()
````

````python
# look ma, ggplot!
ggplot(aes(x='latency', y='value', colour='variable'), data=melted) + \
    geom_line() + \
    scale_y_log10() + \
    xlab('latency ($\mu$s)') + \
    ylab('occurences')
````

The histograms for no real-time and PREEMPT_RT are truncated.

Another way to look at the data is via a cummulative distribution function (CDF).

````python
# another DataFrame wrangling to get the required shape for CDF plot
df1_norm = df1.copy()
df2_norm = df2.copy()
df3_norm = df3.copy()

df1_norm.no_rt = df1_norm.no_rt.cumsum().apply(lambda x: x / df1_norm.no_rt.cumsum().max())
df2_norm.preempt_rt = df2_norm.preempt_rt.cumsum().apply(lambda x: x / df2_norm.preempt_rt.cumsum().max())
df3_norm.rt_xenomai = df3_norm.rt_xenomai.cumsum().apply(lambda x: x / df3_norm.rt_xenomai.cumsum().max())

merged_norm = df1_norm.merge(df2_norm, how='outer', on='latency').merge(df3_norm, how='outer', on='latency')
melted_norm = pd.melt(merged_norm, id_vars='latency')
melted_norm.tail()
````

````python
# CDF plot: probability that latency is less-than-or-equal-to the value in x-axis
# dashed-line: 95% probability cut-off
ggplot(aes(x='latency', y='value', colour='variable'), data=melted_norm) + \
    geom_line() + \
    xlab('x ($\mu$s)') + \
    ylab('Prob ( latency $\leq$ x )') + \
    geom_hline(yintercept=[0.95], color='black', size=2, linetype='dashed') + \
    geom_vline(xintercept=[15], color='black', size=2, linetype='dashed')
````

# Epilogue

Hello World with Xenomai (from [here](http://www.cs.ru.nl/lab/xenomai/exercises/ex01/Exercise-1.html)):

```c
#include <stdio.h>
#include <signal.h>
#include <unistd.h>
#include <sys/mman.h>
 
#include <native/task.h>
#include <native/timer.h>
 
#include  <rtdk.h>
RT_TASK demo_task;
 
void demo(void *arg)
{
  RT_TASK *curtask;
  RT_TASK_INFO curtaskinfo;
 
  // hello world
  rt_printf("Hello World!\n");
 
  // inquire current task
  curtask=rt_task_self();
  rt_task_inquire(curtask,&curtaskinfo);
 
  // print task name
  rt_printf("Task name : %s \n", curtaskinfo.name);
}
 
int main(int argc, char* argv[])
{
  char  str[10] ;
 
  // Perform auto-init of rt_print buffers if the task doesn't do so
  rt_print_auto_init(1);
 
  // Lock memory : avoid memory swapping for this program
  mlockall(MCL_CURRENT|MCL_FUTURE);
 
  rt_printf("start task\n");
 
  /*
   * Arguments: &task,
   *            name,
   *            stack size (0=default),
   *            priority,
   *            mode (FPU, start suspended, ...)
   */
  sprintf(str,"hello");
  rt_task_create(&demo_task, str, 0, 50, 0);
 
  /*
   * Arguments: &task,
   *            task function,
   *            function argument
   */
  rt_task_start(&demo_task, &demo, 0);
}
```

Xenomai is great, but is not easy for "new" applications (i.e. not ported from another RTOS, e.g. VxWorks. Initial goals of Xenomai is to facilitate the porting of programs from traditional RTOS to RTAI on GNU / Linux.), or for applications developed for standard Linux, because it requires using Xenomai's APIs. Also once an application is developed for Xenomai, it will be locked to one certain software choice. Problematic if, say, someday Xenomai is no longer supported, or being acquired by a big bad corporation.

On the other hand, from the CDF curve, even with 100% CPU load, `cyclictest` latency of PREEMPT_RT is $\leq$ 50 $\mu$s for 95% of the time. So it may still work depending on the "real-time" requirement of the applications.
